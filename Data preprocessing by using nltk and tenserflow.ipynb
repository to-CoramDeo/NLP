```python
#데이터 전처리 및 정제 - 1) 토큰화
#사용 되는 라이브러리: Nltk, Tensorflow.keras
#참조: https://wikidocs.net/21698
#<1> 영어권 언어에서 아포스트로피(')가 들어가는 단어는 어떻게 토큰으로 분류할까요?
```

```python
#import를 하여 nltk 패키지(라이브러리)를 사용할 수 있게 합니다.
import nltk
```

```python
nltk.__version__
```

```python
nltk.download('all')
```

```python
#(1) nltk의 tokenize에 있는 word_tokenize를 사용합니다.
from nltk.tokenize import word_tokenize
```

```python
print(word_tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop.")) 
#word_tokenize의 결과
#Don't -> 'Do', "n't"
#Jones's -> 'Jone' + 's'
```

```python
#(2) nltk의 tokenize에 있는 WordPunctTokenizer를 사용합니다.
from nltk.tokenize import WordPunctTokenizer
```

```python
print(WordPunctTokenizer().tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
#WordPunctTokenizer의 결과
#Don't -> Don', "'", 't'
#Jones's -> 'Jone', "'", 's'
#=> 구두점(, . ;)을 별도로 분류하네요.
```

```python
#(3) tensorflow의 keras에 있는 토큰화 도구 text_to_word_sequence를 사용합니다.
from tensorflow.keras.preprocessing.text import text_to_word_sequence
```

```python
print(text_to_word_sequence("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
#text_to_word_sequence의 결과
#Don't -> "don't"
#Jones's -> "jone's"
#=> 1. 모든 알파벳을 소문자로 바꿉니다.
#=> 2. 구두점(, . ; !)을 제거합니다.
#=> 3. 아포스트로피(')를 보존합니다.
```
